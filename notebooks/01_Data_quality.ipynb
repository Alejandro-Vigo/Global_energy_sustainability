```python
import janitor
df = df.clean_names()

print(df.columns)
```

```python
df.columns = (
    df.columns
    .str.strip()          # remove spaces at beginning and end
    .str.replace("\n","") # remove line breaks
    .str.replace(" ","_") # replace spaces with underscores
    .str.lower()          # convert to lowercase
    .str.replace(r"[^\w\d_]+","", regex=True)  # remove special characters
    .str.replace(r'__+', '_', regex=True)  # doubles __ → _
    .str.rstrip('_')  # remove _ final
)

df.columns
```

## Null analysis

```python
df.isna().sum().sort_values(ascending=False)
```

We remove the column renewables_equivalent_primary_energy because it has too many null values and it isn't very important:

```python
df = df.drop(columns=['renewables_equivalent_primary_energy'])
```

We will analyse further the other two variables (financial_flows_to_developing_countries_us, renewable_electricity_generating_capacity_per_capita) with huge amount of null values:        

```python
df[df['financial_flows_to_developing_countries_us'].isna()]
```

```python
df.groupby('year').agg(
    missing_financial=(
        'financial_flows_to_developing_countries_us',
        lambda x: x.isna().mean()
    )
)
```

```python
(
    df['financial_flows_to_developing_countries_us']
    .isna()
    .groupby(df['year'])
    .mean()
    .plot(kind='bar', figsize=(12,4))
)
```

```python
df.groupby('entity').agg(
    missing_financial=(
        'financial_flows_to_developing_countries_us',
        lambda x: x.isna().mean()
    )
).sort_values('missing_financial', ascending=False).head(10)
```

```python
df.groupby('year').agg(
    missing_renewable=(
        'renewable_electricity_generating_capacity_per_capita',
        lambda x: x.isna().mean()
    )
)
```

```python
df.groupby('entity').agg(
    missing_renewable=(
        'renewable_electricity_generating_capacity_per_capita',
        lambda x: x.isna().mean()
    )
).sort_values('missing_renewable', ascending=False).head(10)
```

The variable financial_flows_to_developing_countries_us presents a high proportion of missing values, with missingness heavily concentrated in the year 2020 across a wide range of countries.
The pattern is not random and is likely associated with global reporting disruptions during the COVID-19 pandemic rather than with structural data availability issues.
Given the temporal nature of the missingness, no imputation is performed at this stage.

The variable renewable_electricity_generating_capacity_per_capita exhibits persistent missing values across all years, affecting both developed and developing countries.
The consistency of missingness over time suggests a structural coverage limitation rather than a temporary reporting gap.
The variable is therefore retained with missing values.

## Numerical variables

```python
pd.options.display.float_format = '{:,.2f}'.format
df.describe().T
```

renewable_electricity_generating_capacity_per_capita has the standard deviation 2x more than the mean -> likely presence of outliers.
financial_flows_to_developing_countries_us has large discrepancy -> should not be used without transformation and careful missing-data handling.

Most indicators show high variability, with zeros or extreme values in renewables, nuclear, financial flows, and emissions. Access-related metrics are saturated in many countries, while economic and emissions variables have strong scale effects.
These patterns highlight the need for careful handling of missing values, outliers, and skewness in the next cleaning and preprocessing steps.

**First Insights:**

* Access to electricity and clean fuels is high in most countries, but renewable capacity and financial flows show large gaps.
* Emissions, GDP, and energy consumption are highly skewed, with a few countries dominating totals.
