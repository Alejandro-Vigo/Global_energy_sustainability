```python
df = pd.read_csv('../data/processed/01_df_exploration.csv')
df.shape
```

# Missing values

```python
missing = df.isna().sum()
missing_percent = (missing / len(df) * 100).sort_values(ascending=False)
print(missing_percent)
```

Create a missing flag for tracking:

```python
df['financial_flows_missing'] = df['financial_flows_to_developing_countries_us'].isna().astype(int)
df['renewable_capacity_missing'] = df['renewable_electricity_generating_capacity_per_capita'].isna().astype(int)
```

```python
numeric_vars_country = [
    'value_co2_emissions_kt_by_country',
    'gdp_growth',
    'gdp_per_capita',
    'energy_intensity_level_of_primary_energy_mj_2017_ppp_gdp',
    'renewable_energy_share_in_the_total_final_energy_consumption'
]

for var in numeric_vars_country:
    df[var] = df.groupby('entity')[var].transform(lambda x: x.fillna(x.median()))

df['financial_flows_to_developing_countries_us'] = df.groupby('year')[
    'financial_flows_to_developing_countries_us'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isna().mean().sort_values(ascending=False)
```

Missing values were intentionally preserved for structurally absent or unreported indicators (e.g. renewable capacity, nuclear electricity). Imputation was applied selectively only where appropriate, while key variables retain missingness to avoid introducing artificial signals.
