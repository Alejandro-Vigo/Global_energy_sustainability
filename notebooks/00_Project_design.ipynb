# Global Energy Sustainability Analysis and Optimization

**Description:**

This dataset provides a detailed overview of sustainable energy indicators across nations from 2000 to 2020. It covers critical metrics such as electricity access, renewable energy adoption, carbon emissions, energy intensity, financial flows, and economic growth.

# Objetive

Analyze historical global energy consumption patterns (2000–2020) to determine the correlation between economic growth (GDP) and environmental impact (CO2). The analysis aims to segment countries by their "Energy Maturity" and identify which factors have historically been most effective in driving the transition to sustainable energy.

# Levers

* Financial Investment: Increasing financial flows to developing countries to stimulate green infrastructure.

* Capacity Building: Increasing renewable electricity generating capacity per capita through infrastructure projects.

* Energy Efficiency: Reducing energy intensity level of primary energy (improving technology to use less energy for the same GDP output).

* Policy Focus: Shifting the primary energy consumption mix from fossil fuels to renewables.

# KPIs

* Sustainability: Renewable energy share in total final energy consumption

* Environmental Impact: Value CO2 Emissions

* Social Impact: Access to electricity and Access to clean fuels for cooking

* Economic Correlation: GDP per capita vs. Value CO2 Emissions

# Entities and Data

* Primary Entity: Country

* Time Dimension: Yearly data (2000–2020)

* Dataset Size: 3,649 rows x 21 columns

* Geospatial: Latitude, Longitude.

# Seed Questions

**Economy:** 
* Is there a correlation between GDP per capita and Value CO2 emissions?
* Which countries have managed to increase GDP per capita while decreasing Value CO2 emissions?
* What is the relationship between GDP growth and Renewable electricity? Does investing in green energy slow down economic growth, or is there no negative impact?
* Does international financial aid actually work? Do countries that receive higher Financial flows show a faster increase in Access to electricity or Renewable capacity in the following years?


**Energy Equity & Social Impact:**
* What is the correlation between Access to electricity and GDP per capita? At what GDP threshold does a country typically achieve high access?
* How strongly is Access to clean fuels for cooking correlated with Access to electricity? Do these infrastructures develop simultaneously?


**Geographic & Structural Trends:**
* Is there a significant relationship between Density and Energy intensity level? Are denser countries more energy-efficient per capita?
* How has the total Electricity from fossil fuels vs. Electricity from renewables changed globally from 2000 to 2020?
* Which country has shown the fastest growth rate in Renewable energy share over the last decade?
